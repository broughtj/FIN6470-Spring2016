## Estimating the Hedge Ratio by Linear Regression

<br>
<br>

Let $\Delta S_{t} = S_{t} - S_{t-1}$ and $\Delta F_{t} = F_{t} - F_{t-1}$ be the changes in the spot price at time $t$ and the changes in the futures price at time $t$ respectively. 

<br>

We can estimate the minimum-variance hedge ratio with the following linear regression analysis:

<br>
<br>

$$
\Delta S_{t} = \alpha + \beta \Delta F_{t} + \epsilon_{t}
$$

<br>

We can then estimate the optimal minimum-variance hedge ratio as:

<br>

$$
h^{\ast} = \hat{\beta}
$$

<br>

Let's look into this a bit more. Recall that in the simple linear regression model that we are using here that:

<br>

$$
\begin{aligned}
\beta &= \frac{Cov(\Delta S_{t}, \Delta F_{t})}{Var(\Delta F_{t})} \\
            &= \frac{Corr(\Delta S_{t}, \Delta F_{t}) SD(\Delta S_{t}) SD(\Delta F_{t})}{SD(\Delta F_{t}) SD(\Delta F_{t})} \\
            & = \rho_{s,f} \frac{\sigma_{s} \sigma_{f}}{\sigma_{f} \sigma_{f}}
\end{aligned}
$$

<br>

where $Corr(\Delta S_{t}, \Delta F_{t}) = \rho_{s,f}$ is the correlation coefficient between $\Delta S_{t}$ and $\Delta F_{t}$, and $SD(\Delta S_{t}) = \sigma_{s}$ and $SD(\Delta F_{t}) = \sigma_{f}$ are the standard deviations respectively.

<br>

From this it is easy to see that:

<br>

$$
\hat{h}^{\ast} = \hat{\rho}_{s,f} \frac{\hat{\sigma}_{s}}{\hat{\sigma}_{f}}
$$

<br>

__NB:__ in this setting $\hat{h}^{\ast}$ is assumed to remain constant over the life the of the hedge (_there is no time subscript._)

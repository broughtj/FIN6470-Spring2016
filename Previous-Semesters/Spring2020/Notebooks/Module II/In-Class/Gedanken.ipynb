```julia
function simulate_random_walk(length)
    z = cumsum(randn(length))
    return(z)
end 
```

```julia
y = simulate_random_walk(100);
x = simulate_random_walk(100);
```

```julia
function linregress(y, X)
    β = X \ y
    #yhat = β * X
    #resids = y - yhat
    #return(β, resids)
    return(β)
end
```

```julia
## Run one time
β_hat = linregress(y,x)
```

```julia
## Run many times
M = 10_000_000
stats = zeros(M);

for j in 1:M
    y = simulate_random_walk(100);
    x = simulate_random_walk(100);
    stats[j] = linregress(y,x);
end
```

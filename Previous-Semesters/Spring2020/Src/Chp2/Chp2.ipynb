```python
import numpy as np
import matplotlib.pyplot as plt
```

## Basic Set-Up

```python
S0 = 1000.0
ST = 1050.0
F0 = 1020.0
spot = np.arange(start=0, stop=2000, step=1.0)
longPO = spot - F0
shortPO = F0 - spot
```

## Long Forward Payoff

```python
plt.plot(spot, longPO, linewidth=2, color="orange")
plt.grid()
plt.show()
```

## Short Forward Payoff

```python
plt.plot(spot, shortPO, linewidth=2, color="purple")
plt.grid()
plt.show()
```

## Long and Short Forward Payoffs

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.grid()
ax.set_ylabel("Payoff ($)")
ax.set_xlabel("Spot Price at Maturity ($)")
ax.set_title("Long & Short Forward Payoffs")
ax.plot(spot, longPO, 'orange', spot, shortPO, 'purple')
```
